Load the data, calculate the average alt allele frequency, and plot the histogram of these frequencies

## Load Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.decomposition import PCA 
from sklearn.preprocessing import StandardScaler
```

## Import data 

```python
df_1000_genomes = pd.read_csv("matrix_1kg.txt",
                      sep = '\t')
df_1000_genomes
```

## Generate Alt Counts

```python
average_alt_counts = df_1000_genomes.iloc[:,4:].sum(axis = 1).div(2*(len(df_1000_genomes.columns)-4))
```

## Plot Histogram

```python
fig, ax = plt.subplots()

ax.hist( average_alt_counts )      
# Formatting
ax.set_title('Alelle Frequency Spectrum ')
ax.set_xlabel("Alt. Allele Frequency")
ax.set_ylabel("Count of SNPs")

plt.show()
```
