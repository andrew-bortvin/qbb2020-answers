# Advanced Exercise 1

Plot a time course of FBtr0331261 with each sex in a separate series 

## Load Packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Load raw data

```python
df_fpkms = pd.read_csv( "../../qbb2020/data/fpkms.csv", index_col="t_name" )
df_fpkms
```

## Subset gene of interest

```python
goi = "FBtr0331261"
df_fpkms_subset = df_fpkms.loc[goi, :]
df_fpkms_subset
```

## Plot time course 

```python
x = df_fpkms_subset[1:]

male_bool = []
female_bool = []

for i in range(0,len(x)):
    if x.index[i][0] =='m':
        male_bool.append(True)
        female_bool.append(False)
    else:
        male_bool.append(False)
        female_bool.append(True)

fig, ax = plt.subplots()
ax.plot( x[male_bool].index, x[male_bool] )
ax.set_title('Male Time Course')
ax.set_xlabel("Time")
ax.set_ylabel("Fpkm")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot( x[female_bool].index, x[female_bool] )
ax.set_title('Female Time Course')
ax.set_xlabel("Time")
ax.set_ylabel("Fpkm")
plt.show()
```
